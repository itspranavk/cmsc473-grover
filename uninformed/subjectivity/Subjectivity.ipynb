```python
import json

#open 
f = open("mix5_out.jsonl")
data = json.load(f)
data
```

```python
#original real article
real = data["text"]
real
```

```python
#fake article generated
fake = data["gens_article"][0]
fake
```

```python
#splitting real article by \n
from nltk import tokenize

print("REAL:")
split_real = tokenize.sent_tokenize(real)
split_real
```

```python
#splitting fake article by \n
print("FAKE:")
split_fake = tokenize.sent_tokenize(fake)
split_fake
```

```python
len(split_real)
```

```python
len(split_fake)
```

```python
from textblob import TextBlob
real_res = []
for s in split_real:
    blob = TextBlob(s)
    real_res.append(blob.sentiment.subjectivity)

fake_res = []
for s in split_fake:
    blob = TextBlob(s)
    fake_res.append(blob.sentiment.subjectivity)
```

```python
from statistics import mean

print(mean(real_res))
real_res
```

```python
from statistics import mean

print(mean(fake_res))
fake_res
```

```python
TextBlob(real).sentiment.subjectivity
```

```python
TextBlob(fake).sentiment.subjectivity
```

```python
# sort fake sentences from min subjectivity to max subjectivity
zipped_lists = zip(fake_res, split_fake)

sorted_zipped_lists = sorted(zipped_lists)

sorted_fake = [element for _, element in sorted_zipped_lists]
```

```python
sorted_zipped_lists
```

```python
#inserting sentences according to subjectivity value order from min to max
from numpy.random import default_rng

rng = default_rng()
test_data_raw = [split_real]

for i in range(len(sorted_fake)):
    
    #choose random index(es) for fake sentences to insert at
    indexes_insert = rng.choice(len(split_real)+i, size=i+1, replace=False)
    
    #inserting fake sentences by the order in sorted_fake from min subjectivity to max subjectivity
    curr_mix = split_real.copy()
    for index in range(len(indexes_insert)):
        curr_mix.insert(indexes_insert[index], sorted_fake[index])
    
    
    test_data_raw.append(curr_mix)

#doesn't add the pure fake article to test data set
#test_data_raw.append(split_fake)
    
```

```python
test_data = []
for temp in test_data_raw:
    curr_text = " ".join(temp)
    test_data.append(curr_text)
```
