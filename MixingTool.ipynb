```python
import json

f = open("mix6_out.jsonl")
data = json.load(f)
data
```

```python
#original real article
real = data["text"]
real
```

```python
#fake article generated
fake = data["gens_article"][0]
fake
```

```python
#splitting real article by \n
from nltk import tokenize

print("REAL:")
split_real = tokenize.sent_tokenize(real)
split_real
```

```python
#splitting fake article by \n
print("FAKE:")
split_fake = tokenize.sent_tokenize(fake)
split_fake
```

```python
len(split_real)
```

```python
len(split_fake)
```

```python
from numpy.random import default_rng

rng = default_rng()
test_data_raw = [split_real]

for i in range(len(split_fake)):
    
    #choose i+1 number of random sentences
    indexes_fake = rng.choice(len(split_fake), size=i+1, replace=False)
    #choose random index(es) for fake sentences to insert at
    #if fake article longer than real article, break loop before all sentences in real article are substituted bu fake sentences
    if i+1 >= len(split_real):
        break
    indexes_insert = rng.choice(len(split_real), size=i+1, replace=False)
    
    curr_mix = split_real.copy()
    for index in range(i+1):
        curr_mix[indexes_insert[index]] = split_fake[indexes_fake[index]]
        
    test_data_raw.append(curr_mix)
    
test_data_raw.append(split_fake)
    
```

```python
test_data = []
for temp in test_data_raw:
    curr_text = " ".join(temp)
    test_data.append(curr_text)
```
